```python
import numpy, pandas, seaborn
import matplotlib.pyplot as plt

data_frame = pandas.read_csv('../data/ibm_hr_attrition.csv')
data_frame.info()

plt.figure()
seaborn.catplot(x="Attrition", y="JobSatisfaction", data=data_frame, kind="bar")

plt.figure()
seaborn.boxplot(x="Attrition", y="JobSatisfaction", data=data_frame)

print("Oui, les valeurs sont absurdes")

plt.figure()
seaborn.boxplot(x="JobSatisfaction", y="Attrition", hue="Department", data=data_frame)

seaborn.boxplot(x="Attrition", y="TotalWorkingYears", data=data_frame)


```
